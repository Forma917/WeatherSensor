```python
import csv
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
log_file = "overnight_weather_data.csv"

try:
    with open(log_file,'r') as data:
        all_data=data.readlines()
except:
    print("failed to open csv")
```

```python
weather_data_withheader=[x.strip('\n').split(',') for x in all_data]
header=weather_data_withheader[0]
weather_data=np.asfarray(weather_data_withheader[1:]).T
```

```python
header
```

```python
xdat = 0 #time index
ydat = 3 #temp index

fig,ax=plt.subplots(4,1,figsize=(10,20),dpi=100)

for ydat in range(3,7):
    ax[ydat-3].plot((weather_data[0]-weather_data[0,0])/60,weather_data[ydat])
    ax[ydat-3].set(xlabel=header[xdat]+'(min)',ylabel=header[ydat]);
```
